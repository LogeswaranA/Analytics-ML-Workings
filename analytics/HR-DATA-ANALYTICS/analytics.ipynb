```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
#Read csv
hrdata = pd.read_csv("Human_Resources.csv")
print(hrdata)
```

```python
#Let's realign the key feature variable

hrdata["Attrition"] = hrdata["Attrition"].apply(lambda x: 1 if x=='Yes' else 0)
hrdata["OverTime"] = hrdata["OverTime"].apply(lambda x: 1 if x=='Yes' else 0)
hrdata["Over18"] = hrdata["Over18"].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
print(hrdata[hrdata["Attrition"]==1].head())
```

```python
# let us see, how many employees have left the company
employees_who_left = hrdata[hrdata["Attrition"]==1]
employees_who_stayed = hrdata[hrdata["Attrition"]==0]
```

```python
#Lets count the total of each dattaframe separately
print(len(employees_who_left))
print(len(employees_who_stayed))
```
