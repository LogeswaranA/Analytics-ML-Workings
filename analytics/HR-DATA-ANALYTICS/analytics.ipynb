```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read csv
hrdata = pd.read_csv("Human_Resources.csv")
print(hrdata)
```

```python
pd.set_option('display.max_columns', None)
```

```python
hrdata.info()
```

```python
hrdata.describe()
```

```python
#Let's realign the key feature variable

hrdata["Attrition"] = hrdata["Attrition"].apply(lambda x: 1 if x=='Yes' else 0)
hrdata["OverTime"] = hrdata["OverTime"].apply(lambda x: 1 if x=='Yes' else 0)
hrdata["Over18"] = hrdata["Over18"].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
print(hrdata[hrdata["Attrition"]==1].head())
```

```python
# let us see, how many employees have left the company
employees_who_left = hrdata[hrdata["Attrition"]==1]
employees_who_stayed = hrdata[hrdata["Attrition"]==0]
```

```python
#Lets count the total of each dattaframe separately
print(len(employees_who_left))
print(len(employees_who_stayed))
```

```python
#let see if we have any mising data
sns.heatmap(hrdata.isnull(),yticklabels=False,cbar=False,cmap="Blues")
```

```python
#Graph shows, we do not have missing values which is good!
```

```python
hrdata.hist(bins=30,figsize=(20,20),color="b")
```

```python
#Lets drop unwanted variables which has no impact to dataframe
# Over18, EmployeeCont, StandardHours, EmployeeNumber
hrdata.drop(['Over18','EmployeeCount','StandardHours','EmployeeNumber','BusinessTravel'],axis=1,inplace=True)
```

```python
hrdata.head()
```

```python
#Lets identify the % of people who stayed vs left

print("Employees who left =",1. * len(employees_who_left)/len(hrdata) * 100)
```

```python
print("Employees who stayed =",1. * len(employees_who_stayed) / len(hrdata) * 100 )
```

```python
employees_who_left.describe()
```

```python
employees_who_stayed.describe()
```

```python
#Let's compare both the insights between who left vs stayed
# average age of employee who stayed is greater than who left
# Distance from home average in left dataset is greater than who stayed
# Average monthly income plays significant role.. since average is higher for employees who stayed compared to who left
# Average daily rate is less for employees who left
```

```python
#lets understand the correlation among the variables
correlations = hrdata.corr()
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(correlations,annot=True)

```
